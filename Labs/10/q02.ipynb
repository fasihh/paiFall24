```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import IPython.display as display
```

```python
df = pd.read_csv('adult.csv')

display(df.info())
display(df.head())
df.tail()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 5, 1)
sns.boxplot(df['age'])

plt.subplot(1, 5, 2)
sns.boxplot(df['education.num'])

plt.subplot(1, 5, 3)
sns.boxplot(df['capital.gain'])

plt.subplot(1, 5, 4)
sns.boxplot(df['capital.loss'])

plt.subplot(1, 5, 5)
sns.boxplot(df['hours.per.week'])
plt.tight_layout()
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 5, 1)
sns.histplot(df['age'], kde=True, edgecolor='none')

plt.subplot(1, 5, 2)
sns.histplot(df['education.num'], kde=True, edgecolor='none')

plt.subplot(1, 5, 3)
sns.histplot(df['capital.gain'], kde=True, edgecolor='none')

plt.subplot(1, 5, 4)
sns.histplot(df['capital.loss'], kde=True, edgecolor='none')

plt.subplot(1, 5, 5)
sns.histplot(df['hours.per.week'], kde=True, edgecolor='none')
plt.tight_layout()
```

```python
sns.heatmap(df[['age', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']].corr(), annot=True)
plt.show()
```
