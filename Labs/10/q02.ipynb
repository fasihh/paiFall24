```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import numpy as np
```

```python
df = pd.read_csv('adult.csv')

display(df.info())
display(df.head())
df.tail()
```

```python
df = df.replace('?', df['workclass'].mode()[0])
```

```python
df.describe()
```

```python
plt.figure(figsize=(15, 6))

plt.title('Box plots of all numeric fields', fontsize=20)
for i, field in enumerate(['age', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']):
    plt.subplot(1, 5, i+1)
    sns.boxplot(df[field])
plt.tight_layout()
```

```python
plt.figure(figsize=(20, 5))

for i, field in enumerate(['age', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']):
    plt.title(f'Distribution of {field}')
    sns.histplot(df[field], kde=True, edgecolor='none')
    plt.show()

for feature in ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']:
    # plt.figure(figsize=(12, 6))
    sns.countplot(x=df[feature], order=df[feature].value_counts().index)
    plt.title(f'Distribution of {feature}')
    plt.xticks(rotation=90)
    plt.show()
```

```python
sns.heatmap(df[['age', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']].corr(), annot=True)
```

```python
sns.lineplot(x=df['age'], y=df['hours.per.week'])

plt.figure(figsize=(15, 5))
sns.violinplot(x=df['workclass'], y=df['age'])
plt.show()

sns.countplot(x='education', data=df, hue='income')
plt.xticks(rotation=90)
plt.show()

sns.countplot(x='race', data=df, hue='income')
plt.show()
```

```python
sns.lineplot(x='education.num', y='capital.gain', data=df, color='g')
plt.show()
sns.lineplot(x='education.num', y='hours.per.week', data=df)
plt.show()
```
