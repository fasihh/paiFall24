```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('heart.csv')

plt.figure(figsize=(20, 20))
plt.subplot(4, 4, 1)
sns.histplot(data=df, x='age')
plt.title('Age')
plt.ylabel('Frequency')
plt.xlabel('Value')

plt.subplot(4, 4, 2)
sns.histplot(data=df, x='cp')
plt.title('CP')
plt.ylabel('Frequency')
plt.xlabel('Value')

plt.subplot(4, 4, 3)
sns.histplot(data=df, x='trestbps')
plt.title('TRESTBPS')
plt.ylabel('Frequency')
plt.xlabel('Value')

plt.subplot(4, 4, 4)
sns.histplot(data=df, x='chol')
plt.title('CHOL')
plt.ylabel('Frequency')
plt.xlabel('Value')

plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.histplot(df['chol'], kde=True, edgecolor="none")
plt.ylabel('')
plt.show()
```

```python
genders = df.groupby('sex').size()
plt.pie(genders, labels=['Female', 'Male'], autopct='%.2f')
plt.show()
```

```python
sns.scatterplot(data=df, y='thalach', x='age', hue='target')
plt.show()
```
