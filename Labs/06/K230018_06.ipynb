### Q8

#### a

```python
orders = pd.read_csv('orders.csv')
products = pd.read_csv('products.csv')
orders['order_id'] = orders['order_id'].astype('object')
products['product_id'] = products['product_id'].astype('object')
```

```python
# last 10 of products
products.tail(10)
```

```python
# first 5 of orders
orders.head()
```

```python
# last 10 of orders
orders.tail(10)
```

#### c

```python
transactions = pd.merge(orders, products, 'left')
transactions['total'] = transactions['price'] * transactions['quantity']
f"Total Revenue: {transactions['total'].sum():.2f}"
```

#### d

```python
products_sold = transactions\
    .groupby('product_id')\
    .agg({'quantity': 'sum', 'product_name': 'first', 'category': 'first'})\
    .sort_values('quantity', ascending=False)

# top 5 best selling
products_sold.head()
```

```python
products_sold.tail()
```

#### e

```python
f"Most common category among top 5 best selling: {
    products_sold.head()\
    .groupby('category')\
    .agg({'quantity': 'sum'})\
    .sort_values('quantity', ascending=False)\
    .reset_index()\
    .iloc[0, 0]
}"
```

#### f

```python
transactions.groupby('category').agg({'price': 'mean'})
```

#### g

```python
def get_month(date):
    return date[:2] + '/' + date[6:]

def get_year(date):
    return date[-4:]

transactions['month'] = transactions['order_date (MM-DD-YYYY)'].apply(get_month)
transactions['year'] = transactions['order_date (MM-DD-YYYY)'].apply(get_year)
```

```python
f'Year with max revenue: {transactions.groupby('year').agg({'total': 'sum'}).reset_index().max().iloc[0]}'
```

```python
f'Month with max revenue: {transactions.groupby('month').agg({'total': 'sum'}).reset_index().max().iloc[0]}'
```

```python
f'Date with max revenue: {transactions.groupby('order_date (MM-DD-YYYY)').agg({'total': 'sum'}).reset_index().max().iloc[0]}'
```

#### h

```python
revenue_per_month = transactions.groupby('month').agg({'total': 'sum'})
revenue_per_month
```

#### i

```python
orders.isnull()

def clean(df: pd.DataFrame):
    dtypes = {
        'object': 'NULL',
        'int64': 0,
        'float64': 0,
    }

    for column in df.columns:
        dtype = str(df[column].dtype)
        df[column].fillna(dtypes[dtype])

clean(orders)
clean(products)
```
