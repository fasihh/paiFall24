```python
movies_data = {
    'movie_id': [1, 2, 3, 4, 5],
    'title': ['up', 'downfall', 'look who\'s back', 'jurrasic world', 'jurassic park'],
    'director': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'release_year': [2010, 2004, 2017, 2015, 1997],
    'revenue': [2500000, 3800000, 450000000, 10000000, 7270000],
    'spent': [700000, 1000000, 5000000, 2000000, 300000]
}

movies_df = pd.DataFrame(movies_data).set_index('movie_id')
movies_df[(movies_df['revenue'] > 2000000) & (movies_df['spent'] < 1000000)]
```

```python
movies_df.sort_values(ascending=False, by='revenue')
```
