```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Q1

student_heights = np.random.randint(140, 190, 20)
student_ids = range(20)

# generating 20 different colors
colors = (lambda cm: [cm(i) for i in student_ids])(plt.get_cmap('hsv', 20))

plt.bar(student_ids, student_heights, color=colors)
plt.xlabel('Students')
plt.ylabel('Heights')
plt.show()

heights = dict()
for height in student_heights:
    if height not in heights:
        heights[height] = 1
    else:
        heights[height] += 1

plt.pie(heights.values(), labels=heights.keys(), autopct='%.1f')
plt.show()
```

```python
# Q2

populations = np.random.randint(10000, 50000, 10)

plt.barh([str(i+1) for i in range(10)], populations)
plt.xlabel('Population')
plt.ylabel('Cities')
plt.show()
```

```python
# Q3

# assuming same price for every flavor
flavors = ["Vanilla", "Chocolate", "Strawberry", "Mint", "Cookie Dough", "Pistachio", "Mango"]
scoops_sold = np.random.randint(50, 200, len(flavors))
total_scoops = np.sum(scoops_sold)

plt.pie(scoops_sold, labels=flavors, autopct='%.2f')
plt.show()
```

```python
# Q4

# No dataset provided. Creating my own
n = 500
df = pd.DataFrame({
    "StudentID": range(1, n+1),
    "Name": [f"Student_{i}" for i in range(1, n+1)],
    "Age": np.random.randint(18, 45, n),
    "Gender": np.random.choice(["Male", "Female"], n)
})

age_grouped = pd.cut(df['Age'], bins=[18, 25, 30, 35, 100], labels=['18-25', '26-30', '31-35', '36+']).value_counts()
plt.pie(age_grouped, labels=age_grouped.index, autopct='%.2f')
plt.show()
```

```python
# Q5

gender_group = df.groupby('Gender').size()
plt.pie(gender_group.values, labels=gender_group.index, autopct='%.2f')
plt.show()
```

```python
# Q6

# Write a Python program to draw a scatter plot comparing two subject marks of Mathematics
# and Science. Use marks of 10 students and also add legends for better understanding of the
# data.

maths_marks, science_marks = np.random.randint(0, 100, (2, 10))

plt.scatter(range(10), maths_marks)
plt.scatter(range(10), science_marks)
plt.legend(['Maths', 'Science'])
plt.show()
```

```python
# Q7

df = pd.read_csv('Global_sea_level_rise.csv').rename(columns={'mmfrom1993-2008average': 'Average mm'})

# selecting only 100 years
if len(df) > 100:
    df = df[len(df)-100:]

plt.xlabel('Years (starting from 1923)')
plt.ylabel('Sea level rise (mm)')
plt.scatter(range(len(df)), df['Average mm'])
plt.show()
```
