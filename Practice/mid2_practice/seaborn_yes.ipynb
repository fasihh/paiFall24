```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('StudentsPerformance.csv')
# df.columns = list(map(lambda col: col.title(), df.columns))

df.head()
```

```python
# plt.figure(figsize=(10, 6))

# Scatter plot
sns.scatterplot(data=df, x='math score', y='reading score', hue='gender')
plt.show()

twoxminusone = np.frompyfunc(lambda x: (2 * x) - 1, 1, 1)
twoxminusone(np.array([1, 2, 3, 4, 5]))
```

```python
sns.scatterplot(data=df, x='reading score', y='writing score', style='parental level of education', hue='parental level of education')
plt.show()
```

```python
df['average'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

sns.rugplot(data=df, x='average')
plt.show()
```

```python
sns.histplot(data=df, x='math score', kde=True, edgecolor='none', color='coral', bins=30)
plt.show()
```

```python
sns.countplot(data=df, x='gender', hue='parental level of education')
```

```python
sns.barplot(data=df, x='parental level of education', y='average', estimator="mean")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df, y='math score', x='gender', hue='race/ethnicity', width=0.2, palette='Set3')
```

```python
plt.figure(figsize=(10, 8))
sns.swarmplot(data=df, x='math score', y='gender', hue='lunch', size=4)
```

```python
sns.boxenplot(data=df, y='lunch', x='math score', hue='gender')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', alpha=0.3, hue='gender')
```

```python
sns.pairplot(df)
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='bar', hue='lunch')
```

```python
sns.catplot(data=df, x='gender', y='math score', hue='lunch', kind='violin')
```

```python
sns.catplot(data=df, x='gender', y='math score', hue='lunch', kind='box')
```

```python
sns.catplot(data=df, x='gender', y='math score', hue='lunch', kind='boxen', row='lunch', col='gender')
```

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, linewidth=0.2)
plt.yticks(rotation=0)
plt.show()
```
