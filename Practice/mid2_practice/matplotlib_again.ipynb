```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('StudentsPerformance.csv')

df.head()
```

```python
df['average'] = np.round((df['math score'] + df['reading score'] + df['writing score']) / 3, 2)

gender_count = df['gender'].value_counts()
plt.pie(gender_count.values, autopct=r'%.2f', labels=gender_count.index, colors=['red', 'blue'])
plt.plot()
```

```python
race_counts = df['race/ethnicity'].value_counts()
plt.bar(race_counts.index, race_counts.values, linewidth=[1, 2, 3, 4, 5], edgecolor=['r', 'g', 'b', 'y', 'black'])
plt.xticks(fontsize=12)
plt.show()
```

```python
plt.scatter(df['math score'], df['reading score'], color='coral', alpha=0.4, marker='o')
plt.show()
```

```python
plt.hist(df['average'], [50, 60, 70, 80, 90, 100], cumulative=1, color='coral', alpha=0.5, orientation="horizontal")
plt.show()
```

```python
parental_education_count = df['parental level of education'].value_counts()

explode = np.zeros(len(parental_education_count))
explode[parental_education_count.argmin()] += 0.4

plt.pie(parental_education_count.values, labels=parental_education_count.index, explode=explode, autopct='%.2f%%')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 4, 5, 6]
y2 = [3, 4, 5, 6, 7]

plt.stackplot(x, y1, y2, labels=['first', 'second'])
plt.legend(loc=4)
plt.show()
```

```python
plt.boxplot((df['math score'], df['reading score'], df['writing score']), tick_labels=['math', 'reading', 'writing'], widths=0.5, showmeans=True, patch_artist=True, vert=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to check for prime numbers
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Generate prime numbers and cumulative counts
max_range = 20000
prime_counts = []
cumulative_count = 0

for i in range(max_range + 1):
    if is_prime(i):
        cumulative_count += 1
    prime_counts.append(cumulative_count)

# X values (range) and Y values (cumulative prime counts)
x = np.arange(1, max_range + 1)  # Start from 1 to avoid log(0)
y_primes = prime_counts[1:]  # Adjust for x starting at 1
y_lnx = x / np.log(x)  # y = x / ln(x)

# Create a step plot for cumulative prime counts
plt.step(x, y_primes, label='Cumulative Prime Count', color='blue', where='post')
# Create a plot for x / ln(x)
plt.plot(x, y_lnx, label='x / ln(x)', color='orange')

plt.xlabel('Number')
plt.ylabel('Count')
plt.title('Cumulative Count of Prime Numbers and x / ln(x)')
plt.grid()
plt.legend()
plt.text(727, 0, "wysi", fontdict={ 'color': 'magenta' })
plt.annotate('wyfsi', (727, 727), (1000, 1000), arrowprops=dict(color='pink'), color='magenta')
plt.ylim(0, 3000)  # Adjust the y-axis limit for better comparison
plt.show()
```
